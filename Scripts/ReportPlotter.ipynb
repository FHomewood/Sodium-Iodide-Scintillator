```python
#Import Libraries
import csv
import numpy as num
import pylab as lab
import matplotlib.pyplot as plt
import random

#Define the Filepath and Load .csv
print ("Open csv file called: ")
Readfile = filepath + input()
datafile = open(Readfile + '.csv', 'r')
datareader = csv.reader(datafile, delimiter=',')

#Convert .csv to Two List Arrays
file = []
for row in datareader:
    file.append(row)
channel = [];
count = [];
for row in range(len(file)):
    if row > 6:
        channel.append(float(file[row][0]))
        count.append(float(file[row][2]))

# User Decisions
# Histogram Output (divide by total counts?)
# Calibrated Axis? (channel or estimated energy)
print("what's the title of your graph?")
title = input()
yaxis = r'$Count \ []$'

# Define and Implement Calibration Settings
m = 1.01E+00 # Slope
c = 2.79E+01 # Intercept
channel = [(i-c)/m for i in channel]
xaxis = r'$Energy [keV]$'
    
#Prepare Figure for Plotting
image = plt.figure(num=None, figsize=(20, 6), dpi=500, facecolor='w', edgecolor='k')
plt.grid(True)

plt.plot(channel,count,'k-', linewidth = 0.3) #Plot Spectrum

#Add Custom Formatting
plt.title(title)
plt.xticks(num.arange(0,max(channel),100))
plt.xlabel(xaxis)
plt.ylabel(yaxis)
plt.ylim(0)
plt.xlim(0)
plt.show() # Show

#Save Image File
print("Save png as:")
s = input()
if (s != ""):
    image.savefig(filepath + s)
```
